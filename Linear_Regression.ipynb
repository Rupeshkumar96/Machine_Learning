#### LINEAR REGRESSION ON A CREATED DATASET

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simple dataset
data = {
    'hours': [1, 2, 3, 4, 5, 6],
    'marks': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)

# Visualize
sns.lmplot(x = 'hours', y = 'marks', data=df)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()

# Model
lr = LinearRegression()
lr.fit(df[['hours']], df['marks'])

# Prediction
print("Prediction for 7 hours:", lr.predict([[7]]))

# y = mx + c
print("Slope:", lr.coef_)  # This will calculate the slope
print("Intercept:", lr.intercept_) # This will calculate the intercept
```

#### LINEAR REGRESSION WITH SKLEARN DATASETS

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
X = df[['bmi']]   # 2D (feature)
y = df['target']  # 1D (output)
```

```python
lr = LinearRegression()

lr.fit(X,y)
```

```python
print("Slope (m):", lr.coef_[0])
print("Intercept (c):", lr.intercept_)
```

```python
print("Prediction for BMI = 0.02:", lr.predict([[0.02]]))
print("Prediction for BMI = 0.05:", lr.predict([[0.05]]))
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X))
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.title("Linear Regression: BMI vs Target")
plt.show()
```
