## MULTICOLLINEARITY

```python
# Multicollinearity occurs when two or more independent variables are highly corrleated with each other.
# In a regression model.
# Example:
# If we predict house price using:
# House size
# Number of rooms
# House Area

# Here, House size and House Area may be highly correlated causing multicollinearity.

# # Problems caused:
# 1. Slopes or regression coeffecients becomes unstable.
# 2. Difficult to select features or feature selection becomes difficult.
# 3. Model becomes sensitive some datas changes in the model.

# Detecting multicollinearity
# We use heatmap to detect multicollinearity
df = df.drop('ocean_proximity',axis=1)
```

```python
corr = df.corr()
plt.figure(figsize=[5,5])
sns.heatmap(corr,annot=True)
plt.show()
```

```python
# If corelation is >0.8, 0.9, then multicollinearity likely exists
```

## Variance Inflation Factor (VIF)

```python
# VIF = 1 / 1 - R^2
# If VIF =  1 -> No correlation
# If VIF = 1-5 -> Moderate
# IF VIF > 10 -> Serious multicollinearity
```

```python
# # Handling multicollinaerity
# 1. Remove correlated features
# 2. Use ridge regression
# 3. Use Lasso regression
```

## BIAS VS VARIANCE

#### BIAS

```python
# Errors due to oversimplified assumptions
# Example: fitting a straight line on a  non-linear data
# High Bias -> Model too simple
```

#### Variance


```python
# Error due to model being too sensitive to training data than testing data.
# High Variance -> Model learns noise
```

### FORMULA

```python
# ERROR = (BIAS)^2 + VARIANCE + IRREDUCTIBLE ERRORS
```

```python
# # Visualization Concept
# MODEL TYPE             |     BIAS       |   VARIANCE
# LINEAR REGRESSION      |      HIGH      |     LOW
# POLYNOMIAL REGRESSION  |       LOW      |     HIGH
```
