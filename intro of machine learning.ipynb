# Traditional Programming  vs Machine Learning.

```python
# In Traditional programming ,humans need to write code
# In traditional programming,computer follow rules.

# input + Program = Output

# In machine learning computer  creates rules from data

# INPUT + OUTPUT = LEARNING ALGORITHMS
```

```python
# Take a user input of 5 subject marks
# print the total marks and Percentage
# Grade A if 90+,grade B if 75+ Grade C if 60+,Grade D if 45+ else fail
student_marks={
    'subject':['math','english','java','ml'],
    'marks':[85,75,65,84]
}
# total=0
# for i in range(5):
#     marks=int(input("Enter a number"))
#     total+=marks
if sum(student_marks['marks'])/4>90:
    print("Grade A+")
elif sum(student_marks['marks'])/4>75:
    print('Grade B+')
elif sum(student_marks['marks'])/4>60:
    print('Grade c+')
elif sum(student_marks['marks'])/4>45:
    print('Grade D+')
else:
    print("Fail")
```
