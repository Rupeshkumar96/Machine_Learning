## 1.	Import the dataset using Pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df=pd.read_csv('Kc_house_data.csv')
df.head()
```

## 2.	Check and handle missing values.

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
dtypes=df.select_dtypes(include=np.number)
for num in dtypes:
    q1=df[num].quantile(0.25)
    q3=df[num].quantile(0.75)
    iqr=q3-q1
    ll=q1-1.5*iqr
    ul=q3+1.5*iqr
    outlier=df[(df[num]<ll)|(df[num]>ul)]
    print(f'Total oulier of {num} is :{len(outlier)}')
    df=df[(df[num]>=ll) & (df[num]<=ul)]
```

```python
x=df.drop(['id','date','price'],axis=1)
y=df['price']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```


## 4.	Perform EDA and correlation analysis.





```python
import seaborn as sns
df.drop(['date'],axis=1,inplace=True)
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

## 5.	Apply Simple Linear Regression to predict house price using one feature (e.g., square footage).

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[:5]
```

## 6.	Evaluate model using R², MSE, RMSE.

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
rmse=np.sqrt(mse)
print("Mean Squared Error:",mse)
print("Mean Absolute Error:",mae)
print("r2_score:",r2)
print("Root Mean Squared Error:",rmse)
```
