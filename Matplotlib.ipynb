## Bar Graph
## Funciton Name
## color= it is used to display multiple color
## edgecolor=it is used to the show the color of the border
## linewidth = it is used to increase or decrease the size of the line 
## linesytle = It is use to the change the style of the line
## aplha = It is used to the control the contrast the color
## width = It is used to handle the size of the bar graph
## rotation = It is used to the change the module of the color
## Fontsize = We can resize the font size

```python
x=[10,52,56,48]
y=['car','bike','cycle','train']
```

```python
c=['orange','red','blue','black']
plt.bar(y,x,color=c)
```

```python
c=['orange','red','blue','black']
plt.bar(y,x,color=c,edgecolor='y',linewidth=8,linestyle='--',alpha=0.7,width=0.2,align='edge')
plt.title("Practice on the matplotlic bar graph")
plt.xlabel("Name of the Transport",fontsize=12)
plt.ylabel("Number of the transport",fontsize=12)
plt.show()
```

```python
xx=['car','bike','cycle','train']
yy=[10,52,56,48]
zz=[56,45,78,25]
```

```python
import numpy as np
p=np.arange(len(xx))
width=0.2
p1=[j+width for j in p]
plt.bar(p,yy,color='blue',width=0.2)
plt.bar(p1,zz,color='red',width=0.2)
plt.xticks(p+width/2,xx,rotation=60)
plt.title("Practice on the Matplotlib")
plt.xlabel("Name of the transport",fontsize=12)
plt.ylabel("Number of the transport",fontsize=12)
plt.show()
```

```python
import numpy as np
p=np.arange(len(xx))
width=0.2
p1=[j+width for j in p]
plt.barh(p,yy,color='blue')
plt.barh(p1,zz,color='red')
plt.xticks(p+width/2,xx,rotation=60)
plt.title("Practice on the Matplotlib")
plt.xlabel("Name of the transport",fontsize=12)
plt.ylabel("Number of the transport",fontsize=12)
plt.show()
```
