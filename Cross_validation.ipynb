## CROSS-VALIDATION

```python
# Cross validation
# Problems with train_test_split
# When we split data into 80% training and 20% testing there might be some problems with the data.
# Model performance depends upon how the data was split.
# This may lead to variance in evaluation.
```

#### WHAT IS CROSS VALIDATION.?

```python
# # Cross validation repeatedly splits the dataset and averages the performance.
# # Most common method is:
# # K-FOLD CROSS VALIDATION
# # Example k=5
# # Dataset will split into 5 parts:
# # My_data_values: 1,2,3,4,5
# Fold          Train              Test
# 1             2,3,4,5              1
# 2             1,3,4,5              2
# 3             1,2,4,5              3
# 4             1,2,3,5              4
# 5             1,2,3,4              5

# Final performance = average of all folds
```

#### USE CROSS VALIDATION FOR SMALL DATASETS
#### FOR LARGE DATSETS USE TRAIN_TEST_SPLIT

```python
# Advantages
# Reduces overfitting
# Gives stable model evaluation
# Uses entire datasets and divides into n splits unlike train_test_split
```

#### PRACTICAL USAGE

```python
from sklearn.datasets import load_diabetes
# Cross validation and train_test_split are stored in model_selection
from sklearn.model_selection import cross_val_score 
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
data = load_diabetes()
X = data.data
y = data.target
```

```python
# Create the model
model = LinearRegression()
```

```python
# Apply cross validation (BASELINE CROSS VALIDATION MODEL)

scores = cross_val_score(model,X,y,cv=5,scoring='r2')
# model is the algorithm you want to select
# X,y are training and testing data
# cv are the number of folds
# scoring is the type of eval metrics

print('Cross Validation Scores:',scores)
print('Average scores:',scores.mean())

# instead of one r2 score, we get five r2 scores.
# This will give us a relaibale evaluation
```

```python
# Apply cross validation (Advanced example)

from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)

scores = cross_val_score(ridge,X,y,cv=5)
print(scores)
print('Mean score:',scores.mean())
```
