## 2. Merge : in this the column datatype is same otherwise it throw the error
## It work on the column match


```python
a=pd.DataFrame({
    'key1':[1,2,3,4,5],
    'key2':[10,20,30,4,50],
    'key3':[5,4,3,2,1]
})
b=pd.DataFrame({
    'key1':['a','b','c','d','e'],
    'key5':['m','n','o','p','q'],
    'key6':[9,8,7,6,5]
})
```

```python
a
```

```python
b
```

```python
a['key1']=a['key1'].astype(str)
new=pd.merge(a,b,how='right')
new
```

```python
new=pd.merge(a,b,how='outer')
new
```

```python
new=pd.merge(a,b,how='left')
new
```

```python
new=pd.merge(a,b,how='inner')
new
```

```python
new=pd.merge(a,b,how='cross')
new
```

## 3. join : It work on the index match

```python
## in join the column name is different the use this type
```

```python
a1=pd.DataFrame({
    'key1':[1,2,3,4,5],
    'key2':[10,20,30,4,50],
    'key3':[5,4,3,2,1]},
    index=[1,2,3,4,5]
)
b1=pd.DataFrame({
    'key11':[5,4,5,6,3],
    'key5':[10,20,30,40,50],
    'key6':[9,8,7,6,5]},
    index=[1,2,8,9,7]
)
```

```python
a1
```

```python
b1
```

```python
new = a1.join(b1, how='inner')
print(new)
```

```python
new = a1.join(b1, how='left')
print(new)
```

```python
new = a1.join(b1, how='right')
print(new)
```

```python
new = a1.join(b1, how='outer')
print(new)
```

## if column name is different then we join these type

```python
a1=pd.DataFrame({
    'key1':[1,2,3,4,5],
    'key2':[10,20,30,4,50],
    'key3':[5,4,3,2,1]},
    index=[1,2,3,4,5]
)
b1=pd.DataFrame({
    'key1':[5,4,5,6,3],
    'key5':[10,20,30,40,50],
    'key6':[9,8,7,6,5]},
    index=[1,2,8,9,7]
)
```

```python
new=a1.join(b1,how='right',lsuffix='_left',rsuffix='_right')
new
```
